```python
from tkinter import *
from PIL import ImageTk,Image
import random
from random import choice
from random import shuffle


def open_window():
    top= Toplevel()
    top.title('Image Learning window')
    
   
    my_img1=ImageTk.PhotoImage(file="pics/1.png")
    my_img2=ImageTk.PhotoImage(file="pics/2.png")
    my_img3=ImageTk.PhotoImage(file="pics/3.png")
    my_img4=ImageTk.PhotoImage(file="pics/4.png")
    my_img5=ImageTk.PhotoImage(file="pics/5.png")
    my_img6=ImageTk.PhotoImage(file="pics/6.png")
    my_img7=ImageTk.PhotoImage(file="pics/7.png")
    my_img8=ImageTk.PhotoImage(file="pics/8.png")
    my_img9=ImageTk.PhotoImage(file="pics/9.png")
    my_img10=ImageTk.PhotoImage(file="pics/10.png")
    my_img11=ImageTk.PhotoImage(file="pics/11.png")
    my_img12=ImageTk.PhotoImage(file="pics/12.png")
    my_img13=ImageTk.PhotoImage(file="pics/13.png")
    my_img14=ImageTk.PhotoImage(file="pics/14.png")
    my_img15=ImageTk.PhotoImage(file="pics/15.png")
    my_img16=ImageTk.PhotoImage(file="pics/16.png")
    my_img17=ImageTk.PhotoImage(file="pics/17.png")
    my_img18=ImageTk.PhotoImage(file="pics/18.png")
    my_img19=ImageTk.PhotoImage(file="pics/19.png")
    my_img20=ImageTk.PhotoImage(file="pics/20.png")
    my_img21=ImageTk.PhotoImage(file="pics/21.png")
    my_img22=ImageTk.PhotoImage(file="pics/22.png")
    my_img23=ImageTk.PhotoImage(file="pics/23.png")
    my_img24=ImageTk.PhotoImage(file="pics/24.png")
    my_img25=ImageTk.PhotoImage(file="pics/25.png")
    my_img26=ImageTk.PhotoImage(file="pics/26.png")
    
    
    image_list=[my_img1,my_img2,my_img3,my_img4,my_img5
                ,my_img6,my_img7,my_img8,my_img9,my_img10
                ,my_img11,my_img12,my_img13,my_img14,my_img15
                ,my_img16,my_img17,my_img18,my_img19,my_img20
                ,my_img21,my_img22,my_img23,my_img24,my_img25,my_img26]
 
    my_label= Label(top,image=my_img1)
    my_label.grid(row=0,column=0,columnspan=3)
   

    def forward(image_no):
        global my_label
        global button_forward
        global button_back
        
        
       
        my_label=Label(top,image=image_list[image_no-1])
        button_forward=Button(top,text='>>',command=lambda: forward(image_no+1),bg='black',fg='white')
        button_back=Button(top,text='<<',command=lambda: back(image_no-1),bg='black',fg='white')
        my_label.grid_forget()
    
        if image_no ==26:
            button_forward= Button(top,text='>>',state=DISABLED,disabledforeground='black',bg='black')
    
        my_label.grid(row=0,column=0,columnspan=3)
        button_forward.grid(row=1,column=2)
        button_back.grid(row=1,column=0)
        

    def back(image_no):
        global my_label
        global button_forward
        global button_back
    
        my_label.grid_forget()
        my_label=Label(top,image=image_list[image_no-1])
        button_forward=Button(top,text='>>',command=lambda: forward(image_no+1),bg='black',fg='white')
        button_back=Button(top,text='<<',command=lambda: back(image_no-1),bg='black',fg='white')

        if image_no ==1:
            button_back= Button(top,text='<<',state=DISABLED,bg='black',fg='black')
    
        my_label.grid(row=0,column=0,columnspan=3)
        button_forward.grid(row=1,column=2)
        button_back.grid(row=1,column=0)
       
    
    
    button_back=Button(top,text='<<',command=back,state=DISABLED,disabledforeground='black',bg='black',fg='white')
    button1=Button(top,text='close',command=top.destroy,bg='black',fg='white')
    button_forward=Button(top,text='>>',command=lambda:forward(2),bg='black',fg='white')

    button_back.grid(row=1,column=0)
    button1.grid(row=1,column=1)
    button_forward.grid(row=1,column=2)
    
def open_window1():
    attempts=10
    answer=random.randint(1, 99)
    
    def check_answer():
        global text
        global attempts
        guess=int(entry_window.get())
       
        if answer==guess:
            label9.config(text='Congrats!! you won!.')
            btn_check.pack_forget()
        
        elif attempts == 0:
            label9.config(text='you are out of attempts buddy')
            btn_check.pack_forget()
        
        elif attempts < 0:
            attempt=10
        
        elif guess < answer:
            label9.config(text='Incorrect Guess! Now only '+str(attempts)+' attempts remaining.try big value')
            
        elif guess > answer:
            label9.config(text='Incorect Guess! Now only '+str(attempts)+' attempts remaining.Try a small value')        
        
        attempts=attempts-1
    
        return
    
    top1=Tk()
    top1.title('Number Guessing')
    top1.geometry('300x300')
    
    
    label4=Label(top1,text='Enter the number between 1& 99',bg='blue',fg='white',font=('Algerian',15))
    label4.pack(padx=10,pady=20)

    entry_window=Entry(top1,bg='yellow',width=24,borderwidth=10,fg='black',font=('Algerian',15))
    entry_window.pack(pady=10)

    button_frame=Frame(top1,bg='black')
    button_frame.pack(pady=20)
    
    btn_check=Button(button_frame,text='check',command=check_answer,bg='green',fg='white',font=('Algerian',15))
    btn_check.grid(row=0,column=0,pady=10,padx=10)

    btn_quit=Button(button_frame,text='close',command=top1.destroy,font=('Algerian',15))
    btn_quit.grid(row=0,column=1,pady=10,padx=10)
    
    
    label9=Label(top1,text='You have 10 Attempts, Best of luck!',bg='black',fg='white',font=('Algerian',15))
    label9.pack(pady=10)
    
    top1.geometry('700x300')
    top1.configure(bg='black')
    
    
def open_window2():

    top2=Toplevel()
    top2.title('Alphabet Test')
    top2.geometry('600x350')
    top2.configure(bg='black')

    my_label=Label(top2,text='',font=('Algerian',48),bg='black',fg='white')
    my_label.pack()

    def shuffler():
    
        wordss=['APPLE','BALL','CAT','DUCK','EGG'
            ,'FISH','GOAT','HAT','ICE CREAM','JELLY',
            'KITE','LION','MONKEY','NEST','OWL','PARROT',
            'QUEEN','RABBIT','SUN','TURTLE','UMBRELLA',
            'VIOLIN','WHALE','XYLOPHONE','YOGURT','ZEBRA']
    
        global word
        word=choice(wordss)
    
        break_apart_word=list(word)
        shuffle(break_apart_word)
    
        global shuffled_word
        shuffled_word=''
        for letter in break_apart_word:
            shuffled_word+=letter
        
        my_label.config(text=shuffled_word)
    
    def answer():
        if word==entry_answer.get():
            answer_label.config(text='CORRECT!')
        else:
            answer_label.config(text='INCORRECT!')
        
        
    entry_answer=Entry(top2,font=('Algerian',24),bg='yellow',fg='black',borderwidth=10)  
    entry_answer.pack(pady=20)


    button_frame=Frame(top2,bg='black')
    button_frame.pack(pady=20)


    my_button=Button(button_frame,text='Pick Another Word',command=shuffler,bg='green',fg='white',font=('Algerian',18))
    my_button.grid(row=0,column=0,padx=10)

    answer_button=Button(button_frame,text='Answer',command=answer,bg='green',fg='white',font=('Algerian',18))
    answer_button.grid(row=0,column=1,padx=10)

    close_button=Button(button_frame,text='close',command=top2.destroy,font=('Algerian',18))
    close_button.grid(row=0,column=2,padx=10)

    answer_label=Label(top2,text='',font=('Algerian',18),bg='black',fg='white')
    answer_label.pack(pady=20)

    shuffler()
       
    
root=Tk()
root.title('Learning Game Main Window')

lbl=Label(root,text ='Welcome to Kids Learning Game',bg='blue',fg='white',font=('Algerian',18))   
lbl.pack(padx=0,pady=1)

button=Button(root,text='Learn',command=open_window,bg='red',font=('Algerian',18))
button.pack(padx=15,pady=15)
 
button2=Button(root,text='Guess game',command=open_window1,bg='green',fg='white',font=('Algerian',18))
button2.pack(padx=15,pady=15)

button3=Button(root,text='Test Game',command=open_window2,bg='green',fg='white',font=('Algerian',18))
button3.pack(padx=15,pady=15)

button4=Button(root,text='close',command=root.destroy,font=('Algerian',18))
button4.pack(padx=15,pady=15)

lbl=Label(root,text ='No copyrights.......',bg='black',fg='white',font=('Algerian',18))   
lbl.pack(padx=0,pady=1)


root.configure(bg='black')
root.geometry('300x300')
root.mainloop()
```
